# Grouping the data

We can group the data based on one or more columns using the 'df.groupby()' function. For example, we can use 'df.groupby(['Type 1])['Attack'].mean()' to group the data by the 'Type 1' column and calculate the mean of the 'Attack' column for each group.

```python
# Group the data by the Type 1 column and calculate the mean values of each group for the columns Attack and Defense
type_stats = df.groupby("Type 1")[["Attack", "Defense"]].mean()

# sort the data by the Attack column in descending order
type_stats = type_stats.sort_values("Attack", ascending=False)
print(type_stats.head())

# plot the type_stats
type_stats.plot(kind="bar", figsize=(15, 10), rot=0,fontsize=15, width=0.8)



```

Matplotlib has some pretty cool visualization tools that can be used in conjunction with the Pandas library for visual analysis and understanding.

Below I am creating a radar chart to quickly compare the strengths and weaknesses of each starter pokemon. Charts like these can be used for many different domains.

```python

# This function sets up the radar chart

def radar_factory(num_vars, frame='circle'):
    """
    Create a radar chart with `num_vars` axes.

    This function creates a RadarAxes projection and registers it.

    Parameters
    ----------
    num_vars : int
        Number of variables for radar chart.
    frame : {'circle', 'polygon'}
        Shape of frame surrounding axes.

    """
    # calculate evenly-spaced axis angles
    theta = np.linspace(0, 2*np.pi, num_vars, endpoint=False)

    class RadarTransform(PolarAxes.PolarTransform): # this class is used to transform the data into a radar chart

        def transform_path_non_affine(self, path):
            # Paths with non-unit interpolation steps correspond to gridlines,
            # in which case we force interpolation (to defeat PolarTransform's
            # autoconversion to circular arcs).
            if path._interpolation_steps > 1:
                path = path.interpolated(num_vars)
            return Path(self.transform(path.vertices), path.codes)

    class RadarAxes(PolarAxes): # this class is used to create the radar chart

        name = 'radar' # name of the class
        PolarTransform = RadarTransform # this class is used to transform the data into a radar chart

        def __init__(self, *args, **kwargs): # this function is used to initialize the class
            super().__init__(*args, **kwargs) # this function is used to initialize the class
            # rotate plot such that the first axis is at the top
            self.set_theta_zero_location('N') # this function is used to set the location of the first axis

        def fill(self, *args, closed=True, **kwargs): # this function is used to fill the radar chart
            """Override fill so that line is closed by default""" # this function is used to fill the radar chart
            return super().fill(closed=closed, *args, **kwargs) # this function is used to fill the radar chart

        def plot(self, *args, **kwargs): # this function is used to plot the radar chart
            """Override plot so that line is closed by default"""
            lines = super().plot(*args, **kwargs)
            for line in lines: # this for loop is iterates through the lines in the radar chart
                self._close_line(line) # this function is used to close the lines in the radar chart

        def _close_line(self, line):
            x, y = line.get_data()  # the .get_data() function is used to get the data from the radar chart
            # FIXME: markers at x[0], y[0] get doubled-up
            if x[0] != x[-1]:
                x = np.append(x, x[0]) # this function is used to append the x values to a list
                y = np.append(y, y[0])  # this function is used to append the y values to a list
                line.set_data(x, y) # this function is used to set the data for the radar chart

        def set_varlabels(self, labels): # this function is used to set the labels for the radar chart
            self.set_thetagrids(np.degrees(theta), labels) # this function is used to set the labels for the radar chart

        def _gen_axes_patch(self):
            # The Axes patch must be centered at (0.5, 0.5) and of radius 0.5
            # in axes coordinates.
            if frame == 'circle': # if the frame is a circle
                return Circle((0.5, 0.5), 0.5) # return a circle
            elif frame == 'polygon': # if the frame is a polygon 
                return RegularPolygon((0.5, 0.5), num_vars, # return a polygon
                                      radius=.5, edgecolor="k") 
            else:
                raise ValueError("Unknown value for 'frame': %s" % frame) # if the frame is not a circle or a polygon, raise an error

        def _gen_axes_spines(self): # this function is used to generate the axes spines
            if frame == 'circle': # if the frame is a circle
                return super()._gen_axes_spines() # return the super class
            elif frame == 'polygon': # if the frame is a polygon
                # spine_type must be 'left'/'right'/'top'/'bottom'/'circle'.
                spine = Spine(axes=self, # return a spine
                              spine_type='circle', # return a spine
                              path=Path.unit_regular_polygon(num_vars)) 
                
                


# Defining the data for the Holy Trinity Radar Charts

pokemon_data = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/pandas/master/pokemon_data.csv') # reading in a fresh copy of the pokemon data
pokemon_data = pokemon_data.set_index('#') # setting the index to the pokemon number
pikachu = pokemon_data.loc[25] # getting the data for pikachu
charizard = pokemon_data.loc[6] # getting the data for charizard
squirtle = pokemon_data.loc[7] # getting the data for squirtle

# Define function to create spider chart
def radar_chart(ax, data_list, colors):
    # Get list of feature names and values for each data point
    feature_names = list(data_list[0].keys())[1:] # getting the feature names
    num_features = len(feature_names) # number of features
    for i, data in enumerate(data_list): # this for loop is used to iterate through the data
        values = [data[feature] for feature in feature_names]  # getting the values for each feature
        # Calculate angle for each feature
        angles = [n / float(num_features) * 2 * np.pi for n in range(num_features)] # calculating the angle for each feature 
        angles += angles[:1] # adding the first angle to the end of the list
        values += values[:1] # adding the first value to the end of the list
        # Normalize values
        max_value = max(values)
        values = [float(v)/max_value for v in values]
        # Plot radar chart
        ax.plot(angles, values, color=colors[i], linewidth=2, label=data['Name'])
        ax.fill(angles, values, color=colors[i], alpha=0.25)
    # Set axis limits and ticks
    ax.set_ylim([0, 1])
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(feature_names)
    ax.yaxis.grid(True)
    # Add legend
    ax.legend(loc='upper right', bbox_to_anchor=(1.4, 1.2))


# Create figure and axes
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(16, 6), subplot_kw=dict(polar=True))

# Create list of data for each Pokemon by selecting rows from dataframe



pikachu = {'Name': 'Pikachu', 'HP': 35, 'Attack': 55, 'Defense': 40, 'Sp. Atk': 50, 'Sp. Def': 50, 'Speed': 90}
charizard = {'Name': 'Charizard', 'HP': 78, 'Attack': 84, 'Defense': 78, 'Sp. Atk': 109, 'Sp. Def': 85, 'Speed': 100}
squirtle = {'Name': 'Squirtle', 'HP': 44, 'Attack': 48, 'Defense': 65, 'Sp. Atk': 50, 'Sp. Def': 64, 'Speed': 43}
data = [pikachu, charizard, squirtle]



# Create radar chart for each Pokemon
colors = ['yellow', 'orange', 'teal']
for i, ax in enumerate(axs.flat):
    radar_chart(ax, data[i:i+1], colors[i:i+1])

    # add titles to each subplot
    ax.set_title(data[i]['Name'], position=(0.5, 1.1), ha='center')


plt.show()
```
